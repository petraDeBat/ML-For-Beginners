```python
import matplotlib.pyplot as plt
import pandas as pd
pumpkins = pd.read_csv('../data/US-pumpkins.csv')
pumpkins.head()

pumpkins.isnull().sum()

columns_to_select = ['Package', 'Low Price', 'High Price', 'Date']
pumpkins = pumpkins.loc[:, columns_to_select]
pumpkins
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2
price
month = pd.DatetimeIndex(pumpkins['Date']).month
month

jackolanterns = pd.DataFrame({'Month': month, 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'],'High Price': pumpkins['High Price'], 'Price': price})
jackolanterns

jackolanterns.loc[jackolanterns['Package'].str.contains('1 1/9'), 'Price'] = price/(1 + 1/9)
jackolanterns.loc[jackolanterns['Package'].str.contains('1/2'), 'Price'] = price/(1/2)

price = jackolanterns.Price
month = jackolanterns.Month
plt.scatter(month, price)
plt.show()
```
